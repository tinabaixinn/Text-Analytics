```python
import pandas as pd
```

```python
# read file
df = pd.read_csv("fashion.csv")
df.head()
```

```python
# create output dataframe
data = {'ID': [], 'Positive Score': [], 'Negative Score': [], 'Neutral Score': [], 'Compound': []}
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer #Vader 

sid = SentimentIntensityAnalyzer()
for i in range(0, len(df)):
    sentence = df['review text'][i]
    ss = sid.polarity_scores(sentence)
    data['ID'].append(i)
    data['Negative Score'].append(ss['neg'])
    data['Neutral Score'].append(ss['neu'])
    data['Positive Score'].append(ss['pos'])
    data['Compound'].append(ss['compound'])
    for k in ss:
     print('{0}: {1},  '.format(k, ss[k]), end='')
    print()
```

```python
#write result into csv
output = pd.DataFrame(data=data, columns=['ID', 'Negative Score', 'Neutral Score', 'Positive Score'])
output.to_csv('sentiment_output.csv')
```

```python
df['ID'] = data['ID']
df['Negative Score'] = data['Negative Score']
df['Neutral Score'] = data['Neutral Score']
df['Positive Score'] = data['Positive Score']
df.to_csv('merge_output.csv')
```
